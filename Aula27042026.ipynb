```python
Loop = 0

while Loop <= 100:
  print("loop:",Loop)
  Loop += 1 #incrementar parte que informar coma vai estar trabalhando
          #intervalo
```

```python
#Exemplo 02
contador = 10
while contador > 0:
  print("Contagem", contador)
  contador -= 1 #decrementar parte que informar coma vai estar trabalhando
              #intervalo
```

```python
#Aula 03
soma = 0
numero = 0
while numero != -1:
  numero = int(input("Digite um numero "))
  if numero != -1:
    soma += numero
print("A soma dos números é:", soma)
```

```python
#modelo 04
senha_correta = "1234"
tentativa = ""
while tentativa != senha_correta:
  tentativa = input("Digite a senha: ")
print("Acesso permitido")
```

```python
#Questão 01 Contagem regressica
contador = 10
while contador > 0:
  print("Contagem", contador)
  contador -= 1 #decrementar parte que informar coma vai estar trabalhando
print("Fim da contagem")       #intervalo
```

```python
# Questão 2 Apenas numeros pares
numero = 0

while numero <= 20:
  if numero % 2 == 0: #pares/ 1 impar
    print(numero)
  numero += 2
```

```python
# Questão 3 Soma até -5
soma = 0
numero = 0 # Inicializa com um valor diferente de -5

while numero != -5:
  numero = int(input("Digite um número): "))
  if numero != -5:
    soma += numero

print("A soma dos números digitados é:", soma)
```
